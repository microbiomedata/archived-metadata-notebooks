### Above charts are hard to read, but it looks like there are a max of roughly 7 major clustering. So, look at those.

```python
q = """
WITH TOPVALS AS (
    SELECT *, ROW_NUMBER() 
    over (
        PARTITION BY variable
        order by total desc
    ) AS RowNo 
    FROM totalsdf
)
SELECT * FROM TOPVALS WHERE RowNo <= 7
"""
top7_totalsdf = pysqldf(q)
top7_totalsdf
```

### Becuase of repeating values, and a special "annotated value" column that the concatenation of the variable and value columns
e.g., "ECOSYSTEM" + "Host-associated" -> "ECOSYSTEM: Host-associated"
This helps make the stacked bar chart more readable

```python
def format_eav_value(variable, value):
    if pds.isna(value):
        return value
    else:
        return variable + ": " + str(value)
    
top7_totalsdf['annotated_value'] = top7_totalsdf.apply(lambda row: format_eav_value(row['variable'], row['value']), axis=1)
```

```python
stackeddf = top7_totalsdf[['variable', 'annotated_value', 'total']].pivot(index='variable', columns='annotated_value', values='total')
stackeddf
```

### Display stacked bar chart

```python
sns.set() # use seaborn styles
stackeddf.plot.bar(stacked=True, figsize=(15, 10))
plt.legend(loc='best', bbox_to_anchor=(1, 1), borderaxespad=1)
plt.show()
```

## Build graph of data to explore connections

```python
## create EAV table containing only top 7 values
q = """
select 
    distinct value 
from 
    top7_totalsdf
where
    value is NULL
"""
pysqldf(q)
```

```python
## create EAV table containing only top 7 values
q = """
select 
    *
from 
    eavdf
where value in 
    (select distinct value from top7_totalsdf)
or
    value is NULL -- for now include null values so we can count missing data if needed
"""
top7_eavdf = pysqldf(q)
list(top7_eavdf['value'].unique())
```

### Let's look at the first 10 records (ids 0-9)

```python
tempdf = top7_eavdf[top7_eavdf.id < 10]
g1 = nx.from_pandas_edgelist(tempdf, 'id', 'value', edge_attr='variable', create_using=nx.DiGraph())
edge_labels = nx.get_edge_attributes(g1, 'variable')
# print(edge_labels)
```

```python
plt.figure(figsize=(12, 12))
pos=nx.kamada_kawai_layout(g1)
nx.draw_networkx(g1, with_labels=True, pos=pos)
nx.draw_networkx_edge_labels(g1,with_labels=True, pos=pos)
plt.show() # using plt.show() hides text
```

```python
## draw bipartite graph of g1
x, y = nx.algorithms.bipartite.sets(g1)
pos = nx.drawing.layout.bipartite_layout(g1, x)
nx.draw_networkx(g1, pos=pos, with_labels=True)
```

```python
## another way to create bipartite graph
x = list(tempdf['id']) # get list of id nodes
pos = nx.drawing.layout.bipartite_layout(g1, x)
nx.draw_networkx(g1, pos=pos, with_labels=True)
```

```python
pr = nx.pagerank(g1)
sorted(pr.items(), key=lambda x: x[1], reverse=True)
```

```python
for ix, r in tempdf_pivot.iterrows():
    r.dropna(inplace=True)
    print(dict(r))
```

```python
g2 = nx.DiGraph()
for ix, row in tempdf_pivot.iterrows():
    row.dropna(inplace=True) # drop nan fields from row

    ## create a dict from row so that field names can accessed for each values
    d = dict(row) 
    fields = list(d.keys()) # get field names
    vals = list(d.values()) # get list of values
    
    ## build pairs of values to represent edges in graph
    for i in range(len(vals) - 1):
#         edge = (vals[i], vals[i + 1], label=fields[i + 1])
#         print(edge)
#         g2.add_edge(*edge)
        g2.add_edge(vals[i], vals[i + 1], label=fields[i + 1])
```

```python
plt.figure(figsize=(12, 12))
nx.draw_networkx(g2, with_labels=True, pos=nx.kamada_kawai_layout(g2))
plt.show()
```
